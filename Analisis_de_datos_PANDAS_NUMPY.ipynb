## Análisis de un repertorio de películas 

```python
# libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# conecting to the Database
df = pd.read_csv('https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/peliculas.csv')
df = pd.DataFrame(df)
```

```python
#visualizing the first 5 rows 
df.head(5)
```

```python
# Checking how many rows and columns are there 
df.shape
```

```python
# Number of movies
num_peliculas = df['title'].count()
print('Hay un total de', num_peliculas, 'peliculas')
```

```python
# Number of categories
categorias = df['category'].nunique()
print('El número de categorías es de', categorias)
```

```python
# Average lenght of a movie
print('El tiempo promedio de duración de una pelicula es de',df['length'].mean(), 'minutos')
```

```python
peliculas_dormir = df[(df['length'] > 120) & (df['category'] == 'Documentary')]
```

```python
peliculas_dormir
```

```python
# Number of unique categories
num_categrías = df['category'].nunique()
print('Hay un total de', num_categrías, 'géneros de peliculas')
```

```python
# Number of movies for each category
print('El número de películas por cada género es de')
num_por_categoria = df.value_counts('category')
num_por_categoria = pd.DataFrame(num_por_categoria)
num_por_categoria
```

```python
# Average lenght by category
print('El tiempo promedio para cada categoría de película en minutos es de')
df.groupby('category')['length'].mean().round(2)
```

```python
# Clasification by lenght
df['clasification'] = np.where(df['length']< 90, 'Corta', np.where(df['length']> 120, 'Larga', 'Normal'))
df
```

```python
# Importing the new csv
asgard = pd.read_csv('https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/poblacion_Asgard.csv')
asgard = pd.DataFrame(asgard)
```

```python
# Visualizing the top of the data
asgard.head(5)
```

```python
# Sorting by Estado and Municipio
asgard.sort_values(['Estado', 'Municipio']).head(9)
```

```python
# Cheking the number of Municipios in each Estado
asgard.groupby(['Estado'])[['Municipio']].count()
```

```python
# Printing the number of Municipios in each Estado, the total number of Estados, and the total number of Municipios
num_estados = asgard['Estado'].nunique()
num_munestado = asgard.groupby(['Estado'])[['Municipio']].count()
municipios = asgard['Municipio'].count()
print(f'Hay un total de {num_estados} estados \n')
print('El número de municipios por estado es \n')
print(num_munestado)
print('\nEl número total de municipios es de', municipios)
```

```python
# Asgard's population and its Estado´s population
print('El número total de personas en Asgard es de',asgard['Población'].sum())


p_estado = asgard.groupby(['Estado'])[['Población']].sum()
print(p_estado)
```

```python
# Verifiying Asgard's median population
mediana = asgard['Población'].median()
mediana 
```

```python
# Adding a new column with the clasification and printing it
asgard['Clasificación'] = np.where(asgard['Población'] > mediana, 'Muy Poblados', 'Poco Poblados')
asgard
```

```python
# The top 5 most populated Muncipios in Asgard
print('Los municipios más poblados son')
MasPoblado = asgard[['Municipio','Población']]
MasPoblado.sort_values('Población', ascending = False).head(5)
```

```python
# The least 5 populated Municipios in Asgard
print('Los municipios menos poblados son')
MenosPoblado = asgard[['Municipio','Población']]
MenosPoblado.sort_values('Población', ascending = True).head(5)
```

```python
# The most populated Municipio in each Estado
TopMunicipios = asgard.groupby(['Estado'])[['Municipio','Población']].max()
TopMunicipios = TopMunicipios.sort_values('Población', ascending = False)
TopMunicipios
```

```python
MasPoblado2 = MasPoblado.sort_values('Población', ascending = False).head(15)
MasPoblado2 = pd.DataFrame(MasPoblado2)
```

```python
# Graphic representation of the 15 most populated Municipios
plt.figure(figsize=(10, 5))
plt.bar(MasPoblado2['Municipio'],MasPoblado2['Población'])
plt.title('15 municipios más poblados de Asgard')
plt.xlabel('Municipios')
plt.ylabel('Habitantes')
plt.xticks(rotation=45)
plt.show()
```

```python
# Number of movies by genre

final = df.groupby('category').count().reset_index()
final = final.rename(columns={'category': 'categoría', 'title': 'número de películas'})
final
final = final.sort_values('número de películas', ascending = False)
```

```python
# Graphic representation of the number of movies by genre
plt.figure(figsize=(10, 5))
plt.bar(final['categoría'],final['número de películas'])
plt.title('Número de películas por género')
plt.xlabel('Género')
plt.ylabel('Número de películas')
plt.xticks(rotation=45)
plt.show()
```
